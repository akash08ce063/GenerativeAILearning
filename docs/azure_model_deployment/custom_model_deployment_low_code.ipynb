## Model Deployment via Endpoint

The important concept needs to understand for deploying a mode to endpoint is "score.py" file, this file mainly contains two method first, init() and second, run(raw_data) method. First method, initialize all the model related stuff like HuggingFace AutoCasualML class and then each time endpoint hits it call run method.

"Enviornment" -> so to run this score.py file you need to install relevant packages so, conda.yml file usually use to install dependencies. 


Now the endpoint is superb concept, so at each endpoint you can place multiple model as deployment, and you can decide the route the traffic based on percentage.
